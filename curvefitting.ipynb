```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
x = np.array([0.24, 0.65, 0.95, 1.24, 1.73, 2.01, 2.23, 2.52])
y = np.array([0.23, -0.23, -1.10, -0.45, 0.27, 0.10, -0.29, 0.24])
```

```python
# Linear: y = m x + c
coeff_linear = np.polyfit(x, y, 1)
m, c = coeff_linear

print("Linear Fit:")
print(f"y = {m:.6f} x + {c:.6f}")
```

```python
# Quadratic: y = ax^2 + bx + c
coeff_quad = np.polyfit(x, y, 2)
a2, b2, c2 = coeff_quad

print("Quadratic Fit:")
print(f"y = {a2:.6f} x^2 + {b2:.6f} x + {c2:.6f}")
```

```python
degree = 5
coeff_poly = np.polyfit(x, y, degree)

print(f"Polynomial degree {degree} coefficients:")
print(coeff_poly)
```

```python
X = np.linspace(min(x), max(x), 300)

y_linear = np.polyval(coeff_linear, X)
y_quad   = np.polyval(coeff_quad, X)
y_poly   = np.polyval(coeff_poly, X)

plt.figure(figsize=(10,6))
plt.scatter(x, y, color='black', label='Data Points')
plt.plot(X, y_linear, label='Linear Fit')
plt.plot(X, y_quad, label='Quadratic Fit')
plt.plot(X, y_poly, label=f'Polynomial (deg={degree})')

plt.legend()
plt.grid(True)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear, Quadratic, Polynomial Fits")
plt.show()
```

```python
def f_model(x, a, b, c):
    return a * np.log(x) + b * np.cos(x) + c * np.exp(x)
```

```python
popt, pcov = curve_fit(f_model, x, y, p0=[1, 1, 1])  # initial guess
a, b, c = popt

print("Nonlinear Fit:")
print(f"a = {a:.6f}")
print(f"b = {b:.6f}")
print(f"c = {c:.6f}")
```

```python
y_nonlinear = f_model(X, a, b, c)

plt.figure(figsize=(10,6))
plt.scatter(x, y, color='black', label='Data Points')
plt.plot(X, y_nonlinear, label='Nonlinear Fit', linewidth=2)

plt.legend()
plt.grid(True)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Nonlinear Fit: a ln(x) + b cos(x) + c e^x")
plt.show()
```

```python
Phi = np.sum((f_model(x, a, b, c) - y)**2)
print("Phi(a,b,c) =", Phi)
```
